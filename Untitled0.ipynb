```python
import fitz  # PyMuPDF
import requests
import tempfile
import os

def extract_text_from_pdf(pdf_url):
    # Download the PDF
    response = requests.get(pdf_url)
    response.raise_for_status()

    # Save to a temporary file
    with tempfile.NamedTemporaryFile(delete=False, suffix=".pdf") as tmp:
        tmp.write(response.content)
        tmp_path = tmp.name

    # Open with PyMuPDF
    doc = fitz.open(tmp_path)
    text = "\n".join([page.get_text() for page in doc])
    doc.close()

    # Clean up temp file
    os.remove(tmp_path)

    return text
```

```python
# ✅ Correct usage
pdf_text = extract_text_from_pdf(
    "https://raw.githubusercontent.com/Dhihram/RAG_book_sample/main/baliguide-sample-pdf.pdf"
)

print(pdf_text[:1000])  # Preview only
```

```python
from langchain_text_splitters import CharacterTextSplitter

splitter = CharacterTextSplitter(
    separator="\n",
    chunk_size=1000,
    chunk_overlap=200,
)

chunks = splitter.split_text(pdf_text)
```

```python
from langchain_core.documents import Document

chunk_documents = [Document(page_content=chunk) for chunk in chunks]
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings

embedding_model = GoogleGenerativeAIEmbeddings(model="models/gemini-embedding-001")
```

```python
from faiss import IndexFlatL2
from langchain_community.docstore.in_memory import InMemoryDocstore
from langchain_community.vectorstores import FAISS

index = IndexFlatL2(
    len(embedding_model.embed_query("hello world"))
)
vector_db = FAISS(
    embedding_function=embedding_model,
    index=index,
    docstore=InMemoryDocstore(),
    index_to_docstore_id={},
)

# Add the chunk_documents to the vector_db
vector_db.add_documents(chunk_documents)
```

```python
from langchain_groq import ChatGroq

llm = ChatGroq(
    model="llama3-8b-8192",  # Fast & accurate
    temperature=0
)
```

```python
QUESTION = """
How can I commute to Uluwatu in Bali?
"""
```

```python
relevant_docs = vector_db.similarity_search(QUESTION, k=3)
```
